```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/muham/Downloads/company_sales_data.csv")
df.head()
```

```python
month_data = df['month_number']
total_profit_data = df ['total_profit']
bathing_soap_data = df['bathingsoap']
facewash_data = df['facewash']
facecream_data = df['facecream']
```

```python
face_products = facecream_data + facewash_data

# line plot
plt.plot(month_data, face_products, marker = 0, color ='b', linestyle= '--')
plt.title('Face Products Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit For Face products')
plt.grid(True)
plt.show()
```

```python

# line plot
plt.plot(month_data, total_profit_data, marker = 0, color ='b', linestyle= '--')
plt.title('Company Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
fig ,axis =plt.subplots(2, 1, figsize=(8, 6))

axis[0].plot(month_data, bathing_soap_data, marker='o', color='green', linestyle='--') 
axis[0].set_title('Bathing Saop Sales data')
axis[0].set_xlabel('Month Number')
axis[0].set_ylabel('Bathing Soap Sales Unit')
axis[0].grid(True)

axis[1].plot(month_data, facewash_data, marker='o', color='yellow', linestyle='--') 
axis[1].set_title('Face Wash Sales Data')
axis[1].set_xlabel('Month Number')
axis[1].set_ylabel('Face Wash Sales Unit')
axis[1].grid(True)

plt.tight_layout()
plt.show()
```
