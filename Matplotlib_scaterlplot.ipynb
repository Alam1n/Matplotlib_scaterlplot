```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Sample dataset
np.random.seed(42)
x = np.arange(0, 10, 0.5)
y = np.sin(x)
random_data = np.random.normal(loc=5, scale=2, size=100)
print(random_data)
```

```python
# ---- Matplotlib: Line and Scatter ----
plt.figure(figsize=(10, 5))
```

```python
# Line Plot
plt.subplot(1, 2, 1)
plt.plot(x, y, label="Sine Wave", color="blue")
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

# Scatter Plot
plt.subplot(1, 2, 2)
plt.scatter(x, y, color="red", label="Sine Dots")
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# ---- Seaborn: Histogram and Box Plot ----
df = pd.DataFrame({"Random Data": random_data})

plt.figure(figsize=(10, 5))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(df["Random Data"], kde=True, color="green")
plt.title("Histogram")
```

```python
# Box Plot
plt.subplot(1, 2, 2)
sns.boxplot(data=df, orient="h")
plt.title("Box Plot")

plt.tight_layout()
plt.show()
```
